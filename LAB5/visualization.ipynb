```python
import json
import matplotlib.pyplot as plt
import os

# Ścieżki do JSONów
dense_path = "artifacts/metrics_dense.json"
cnn_path = "artifacts/metrics_cnn.json"

def load_metrics(path):
    with open(path, "r") as f:
        return json.load(f)

dense_metrics = load_metrics(dense_path)
cnn_metrics = load_metrics(cnn_path)

dense_metrics, cnn_metrics
```

```python
print("Dense model:")
print("Accuracy:", dense_metrics["test_accuracy"])
print("Loss:", dense_metrics["test_loss"])
print("\nCNN model:")
print("Accuracy:", cnn_metrics["test_accuracy"])
print("Loss:", cnn_metrics["test_loss"])
```

```python
import numpy as np
import seaborn as sns

dense_cm = np.array(dense_metrics["confusion_matrix"])
cnn_cm = np.array(cnn_metrics["confusion_matrix"])

plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
sns.heatmap(dense_cm, annot=False, cmap="Blues")
plt.title("Dense Confusion Matrix")

plt.subplot(1,2,2)
sns.heatmap(cnn_cm, annot=False, cmap="Greens")
plt.title("CNN Confusion Matrix")

plt.tight_layout()
plt.show()
```

```python
def plot_history(history, title):
    epochs = range(1, len(history["loss"]) + 1)

    plt.figure(figsize=(8,5))
    plt.plot(epochs, history["loss"], label="Training Loss")
    plt.plot(epochs, history["val_loss"], label="Validation Loss")
    plt.title(f"{title} – Loss Curve")
    plt.xlabel("Epochs")
    plt.ylabel("Loss")
    plt.legend()
    plt.grid(True)
    plt.show()

    plt.figure(figsize=(8,5))
    plt.plot(epochs, history["accuracy"], label="Training Accuracy")
    plt.plot(epochs, history["val_accuracy"], label="Validation Accuracy")
    plt.title(f"{title} – Accuracy Curve")
    plt.xlabel("Epochs")
    plt.ylabel("Accuracy")
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
plot_history(dense_metrics, "Dense Model")
plot_history(cnn_metrics, "CNN Model")
```
